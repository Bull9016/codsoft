```python
#importing all required liabraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load And explore dataSet
# Load the dataset
data = pd.read_csv('Titanic-Dataset.csv')
```

```python
# Summary statistics and missing values analysis
print("\nSummary of the dataset:")
print(data.describe(include='all'))

print("\nMissing values in each column:")
print(data.isnull().sum())
```

```python
# Visualizing missing values
plt.figure(figsize=(10, 6))
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Distribution of the target variable (Survived)
sns.countplot(data=data, x='Survived',hue='Survived', palette='Set2',legend=False)
plt.title('Survival Count')
plt.show()
```

```python
# Gender distribution and survival rate by gender
sns.countplot(data=data, x='Sex', hue='Sex', palette='cool', legend=False)
plt.title('Gender Distribution')
plt.show()

sns.barplot(data=data, x='Sex', y='Survived', hue='Sex', palette='cool', legend=False)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=30, kde=True, color='blue')
plt.title('Age Distribution')
plt.show()
```

```python
# Survival rate by passenger class (Pclass)
sns.countplot(data=data, x='Pclass', hue='Survived', palette='Set1')
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
# Fare distribution
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Survived', y='Fare',hue='Survived', palette='Set3',legend=False)
plt.title('Fare Distribution by Survival')
plt.show()
```

```python
# Embarkation port distribution and survival rate by Embarkation port
sns.countplot(data=data, x='Embarked',hue='Embarked', palette='viridis',legend=False)
plt.title('Embarkation Port Distribution')
plt.show()
```

```python
sns.barplot(data=data, x='Embarked', y='Survived',hue='Embarked', palette='viridis',legend=False)
plt.title('Survival Rate by Embarkation Port')
plt.show()
```
